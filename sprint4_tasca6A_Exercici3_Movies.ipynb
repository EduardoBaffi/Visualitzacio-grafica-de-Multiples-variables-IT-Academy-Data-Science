```python
# Reading as .csv
movies_df = pd.read_csv('C:\\Users\\eduba\\Documents\\Notebooks\\movies.dat', sep='::', header=None, names=["Name", "Genre"], encoding='ISO-8859-1', warn_bad_lines = True, engine='python')
movies_df.head()
```

```python
# Creación de una columna con solamente el nombre de las película.
new = movies_df["Name"].str.split("\(\d{4}\)", n =-1, expand = True)
movies_df['Movie']=new[0]

# Creación de una columna con solamente el año de las película en separado de los nombres de la película.
movies_df["Year"] = movies_df["Name"].str.extract("\((\d{4})\)")

# Exclir la columna 'Name' porque ya tenemos la misma información en columnas nuevas
movies_df.drop(columns=["Name"], inplace = True)

movies_newdf = movies_df.copy()
movies_newdf = movies_newdf[["Movie","Year","Genre"]]
movies_newdf.head()
```

```python
# Creación de las columnas de cada género con valores booleanos para cada película. 
movies_newdf['Drama'] = movies_newdf['Genre'].str.contains('Drama')
movies_newdf['Romance'] = movies_newdf['Genre'].str.contains('Romance')
movies_newdf['Thriller'] = movies_newdf['Genre'].str.contains('Thriller')
movies_newdf['Childrens'] = movies_newdf['Genre'].str.contains('Children')
movies_newdf['Sci-Fi'] = movies_newdf['Genre'].str.contains('Sci-Fi')
movies_newdf['Comedy'] = movies_newdf['Genre'].str.contains('Comedy')
movies_newdf['Action'] = movies_newdf['Genre'].str.contains('Action')
movies_newdf['Adventure'] = movies_newdf['Genre'].str.contains('Adventure')
movies_newdf['War'] = movies_newdf['Genre'].str.contains('War')
movies_newdf['Horror'] = movies_newdf['Genre'].str.contains('Horror')
movies_newdf['Crime'] = movies_df['Genre'].str.contains('Crime')
movies_newdf['Mystery'] = movies_newdf['Genre'].str.contains('Mystery')
movies_newdf['Musical'] = movies_newdf['Genre'].str.contains('Musical')
movies_newdf['Fantasy'] = movies_newdf['Genre'].str.contains('Fantasy')
movies_newdf['Western'] = movies_newdf['Genre'].str.contains('Western')
movies_newdf['Film-Noir'] = movies_newdf['Genre'].str.contains('Film-Noir')
movies_newdf['Animation'] = movies_newdf['Genre'].str.contains('Animation')
movies_newdf['Documentary'] = movies_newdf['Genre'].str.contains('Documentary')
```

```python
warnings.filterwarnings('ignore')
movie_sum = movies_newdf.groupby("Year",as_index=False)['Drama','Romance', 'Thriller', 'Childrens',
       'Sci-Fi', 'Comedy', 'Adventure', 'War', 'Horror', 'Crime', 'Mystery',
       'Musical', 'Fantasy', 'Western', 'Film-Noir', 'Animation',
       'Documentary', 'Action'].sum()
movie_sum.head()
```

```python
#Lollipop plot del total de peliculas por género cinematográfico
ordered_df = genres_total.sort_values(by='Total')
my_range=range(0,len(genres_total.index))

plt.stem(ordered_df['Total'])
plt.xticks(my_range, ordered_df['Genre'], rotation=90)
plt.show()
```

```python
# Circular barplot del total de peliculas por género cinematográfico

plt.figure(figsize=(26,12))

ax = plt.subplot(111, polar=True)

plt.axis('off')

upperLimit = 150
lowerLimit = 50

max = ordered_df['Total'].max()

slope = (max - lowerLimit) / max
heights = slope * ordered_df.Total + lowerLimit

width = 2*np.pi / len(ordered_df.index)

indexes = list(range(1, len(ordered_df.index)+1))
angles = [element * width for element in indexes]
angles

# Draw bars
bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="white",
    color="steelblue",)

labelPadding = 4

# Add labels
for bar, angle, height, label in zip(bars,angles, heights, ordered_df['Genre']):

   
    rotation = np.rad2deg(angle)
  
    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"

    ax.text(
        x=angle, 
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor") 
```

```python
genres = movies_newdf.copy()
genres.drop(columns=["Genre","Movie","Year"], inplace = True)
genres.head()
```

```python
genre_relations = pd.DataFrame()
for type in genres.columns:
    detail = genres[genres[type] == 1].apply(np.sum)
    detail.rename(type, inplace=True)
    genre_relations = genre_relations.append(detail)
pd.set_option('precision', 0)
genre_relations
```

```python
# Contour plot
sns.kdeplot(x=genre_relations['Comedy'], y=genre_relations['Drama'], cmap="YlGn", shade=True, bw_adjust=.5)
#plt.savefig('tasca6A_ex3_fig1.png', bbox_inches='tight')
#plt.show()
```

```python
# Heatmap de los generos y su relación con los otros Géneros
plt.figure(figsize=(20,10))
sns.heatmap(genre_relations, annot=True, cmap="icefire", annot_kws={"size": 12}, fmt='.0f', robust=True, linewidths=2, linecolor='gray')
plt.title("Relations between Genres")
```

```python
movie_sum["Year"]= movie_sum["Year"].astype(int)
```

```python
decade = (10 * (movie_sum["Year"] // 10)).astype(str) + 's'
movie_sum.insert(0, "Decade",decade)
movie_sum.head()
```

```python
movies_decade = movie_sum.groupby(["Decade"],as_index=False).sum()
movies_decade.drop('Year', 1, inplace = True)
movies_decade
```

```python
movies_decade.info()
```

```python
sample_moviesdec = movies_decade[["Decade","Drama","Comedy","Action","Childrens"]]
sample_moviesdec
```

```python
# Pairplot de algunos generos en cada decada
sns.pairplot(sample_moviesdec, kind="scatter", hue="Decade", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```

```python
# Parallel Coordinates - Evolución de loa gérenos en las décadas
from pandas.plotting import parallel_coordinates

parallel_coordinates(movies_decade, 'Decade', colormap=plt.get_cmap("Set2"))
plt.show()
```

```python
plt.figure(figsize=(20,10))

df = movies_decade.pivot_table(index="Decade", values = ['Drama', 'Romance', 'Thriller', 'Childrens', 'Sci-Fi',
       'Comedy', 'Adventure', 'War', 'Horror', 'Crime', 'Mystery', 'Musical',
       'Fantasy', 'Western', 'Film-Noir', 'Animation', 'Documentary',
       'Action'])
sns.heatmap(df, cmap="Blues", robust=True, linewidths=2, linecolor='white')
plt.title("Number of movies per Genre vs Decade")
```
